```python
import numpy as np
import pandas as pd
import datetime
import random
import itertools as it


#Gather User-Input
year = input("Enter Year: ")
month = input("Enter Month: ")
day = input("Enter Day: ")
td = input("Days Between Games: ")
nTeams = input("Enter Number of Teams: ")
nGames = nTeams

#Turn Str values to Int values
year = int(year)
month = int(month)
day = int(day)
bd = int(td)
nGames = int(nGames)
nTeams = int(nTeams)

#Use the User-Input to create the Starting Date (sd)
sd = datetime.date(year,month,day)

#Display the Start Date and Number of Games
print('Start Date:',sd)
print('Number of Games:',nGames)

#Use User-Input to create a timedelta variable for Days Between Games
tdelta = datetime.timedelta(days=bd)

#User-Input ALL teams
tms = []
while len(tms)<nTeams:
    add_team = input("Add Team: ")
    tms.append(add_team)
```

```python
#Adding ALL Game Dates into 'd' variable
import math as math

d = []
gamesPerWeek = math.floor(nTeams/2)
gamesPerWeek = int(gamesPerWeek)
totalGames = gamesPerWeek*nGames
totalGames = int(totalGames)

def game_dates():
    while len(d)<nGames:
        add_game = sd+(tdelta*len(d))
        d.append(add_game)

for _ in range(gamesPerWeek):
    game_dates()

d
```

```python
from itertools import combinations

pairings = list(it.permutations(tms,2))
```

```python
#use to randomize home and away
from random import shuffle

newpairs = [(i) for i in pairings]
finalpairs = random.sample(newpairs, k=totalGames)

finalpairs
```

```python
tmsH = []
while len(tmsH)<totalGames:
    add_HomeTeam = finalpairs[len(tmsH)][0]
    tmsH.append(add_HomeTeam)
    
tmsA = []
while len(tmsA)<totalGames:
    add_AwayTeam = finalpairs[len(tmsA)][1]
    tmsA.append(add_AwayTeam)
```

```python
#SCHEDULE GENERATOR    
from random import shuffle

gameList = list(range(1,(totalGames+1)))

#Generate Schedule
schedule = pd.DataFrame({'Date': (d * gamesPerWeek), 'Home': (tmsH), 'Away': (tmsA)})
schedule.sort_values(by=['Date'], inplace=True, ascending=True)
schedule.index = gameList
schedule
```

```python
schedule.to_csv('schedule_example.csv')
```
